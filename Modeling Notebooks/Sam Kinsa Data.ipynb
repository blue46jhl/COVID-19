```python
url = 'https://static.kinsahealth.com/US_data.json'
response = requests.get(url).json()
```

```python
data = pd.DataFrame(response['data'], columns = response['columns'])
```

```python
np.where(data['atypical_ili'].isna()==False)
```

```python
fig, ax = plt.subplots()
ax.plot(data['date'], data['atypical_ili'])
for index, label in enumerate(ax.xaxis.get_ticklabels()):
    if index % 6 != 0:
        label.set_visible(False)
ax.set_title("Atypical Fever Temperature: U.S")
```
